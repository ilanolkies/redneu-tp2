```python
data, labels = load_dataset('../dataset/tp2_training_dataset.csv')

data_train = data[:675] # 75% de 900
labels_train = labels[:675]

data_test = data[675:]
labels_test = labels[675:]
```

```python
modelo = Hebbiano(data_train)
```

```python
y = modelo.test(data_train)
plot(y, labels_train)
```

```python
y = modelo.test(data_test)
plot(y, labels_test, 5)
```
